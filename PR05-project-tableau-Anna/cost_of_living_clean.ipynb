### CSV 2 we'll use

```python
# open csv & # transforme data so we have the rows as cities(location)
cost_living_clean = pd.read_csv('cost-of-living.csv', index_col=[0]).T.reset_index()
```

```python
cost_living_clean
```

```python
# data needs to be transformed so we have the rows are cities.
cost_living_clean = cost_living_clean.rename(columns={'index':'location'})
cost_living_clean.head()
```

```python
# separamos de la columna location Country y City
cost_living_clean["Country"]=cost_living_clean["location"].str.split(",").str[-1]
```

```python
cost_living_clean["City"]=cost_living_clean["location"].str.split(",").str[0]
```

```python
# save new csv as cost_living_clean
cost_living_clean.to_csv('cost_living_clean.csv')
```
